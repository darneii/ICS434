#### 2. Generating a scatter plot
* Generate random data to produce a scatter plot
  * You can use any function to generate the data
  * Instead of using circles as symbols, you should use squares.
    

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
students_affiliations = np.random.randint(1, 1000, 4)
x = np.array(["Physics","CS","Math","Engineering"])
```

```python
plt.xlabel("Disciplines")
plt.ylabel("Counts")
plt.title("Distributions of Students in the Program")
_ = plt.bar(x,students_affiliations, color="#FF5733", alpha=0.55)
```

```python
age = np.random.randint(10, 80, 100)
games = np.random.randint(0, 100, 100)
```

```python
plt.xlabel("Age")
plt.ylabel("Games Played")
plt.title("Representation of Participant Age and Amount of Games Played")
_ = plt.scatter(age, games, marker="s")
```
